# QUESTION NO 20

```python
Q20=pd.read_csv(r"C:\Users\VAISHANVI\Downloads\Cars.csv")
Q20
```

```python
from scipy import stats
prob_MPG_is_greater_than_38 = np.round( 1- stats.norm.cdf(38,loc= Q20.MPG.mean(), scale= Q20.MPG.std()),3)
print('P(MPG>38)=',prob_MPG_is_greater_than_38)
```

```python
prob_MPG_is_less_than_40 = np.round(stats.norm.cdf(40,loc= Q20.MPG.mean(), scale= Q20.MPG.std()),3)
print('P(MPG<40)=',prob_MPG_is_less_than_40)
```

```python
prob_MPG_is_less_than_50 = np.round( stats.norm.cdf(50,loc= Q20.MPG.mean(), scale= Q20.MPG.std()),3)
print('P(MPG<50)=',prob_MPG_is_less_than_50)
```

```python
prob_MPG_is_greater_than_20 = np.round(1- stats.norm.cdf(20,loc= Q20.MPG.mean(), scale= Q20.MPG.std()),3)
print('P(MPG>20)=',prob_MPG_is_greater_than_20)
```

```python
prob_MPG_is_greater_than_20_and_less_than_50=(prob_MPG_is_less_than_50)-(prob_MPG_is_greater_than_20)
print('20<P(MPG>50)=',prob_MPG_is_greater_than_20_and_less_than_50)
```
